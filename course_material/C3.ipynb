```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
sentences = [
    'I love my dog',
    'I love my cat',
]
```

```python
tokenizer = Tokenizer(num_words=100)
```

```python
tokenizer.fit_on_texts(sentences)
```

```python
word_index = tokenizer.word_index
```

```python
print(word_index)
```
