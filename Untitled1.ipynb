```python
import tensorflow as tf
```

```python
dataset = tf.data.Dataset.range(10)
```

```python
for val in dataset:
    print(val)
```

```python
for val in dataset:
    print(val.numpy())
```

```python
dataset = tf.data.Dataset.range(10)
for val in dataset:
    print(val.numpy())
```

```python
dataset = dataset.window(3, shift=1)
for window_dataset in dataset:
    print(window_dataset)
```

```python
for window_dataset in dataset:
    print([item.numpy() for item in window_dataset])
```
